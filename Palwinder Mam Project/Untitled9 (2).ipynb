```python
# ============================================================
# CASB-IRL COMPLETE UNIFIED FRAMEWORK
# ============================================================
# FEATURES:
# ✔ Context-Aware Student Behavior IRL
# ✔ Dual Reward Context Modeling
# ✔ Student-Specific Reward Adaptation
# ✔ Personalized Reward Function Learning
# ✔ Behavioral Impact Analysis
# ✔ Local Reward Attribution
# ✔ Explainable AI (XAI)
# ✔ SHAP Dashboard
# ✔ Causal Behavioral Reasoning
# ✔ Excel Explainability Report
# ✔ Google Colab Ready
# ============================================================

# ============================================================
# INSTALL LIBRARIES
# ============================================================

!pip uninstall -y shap -q

!pip install shap==0.46.0 -q
!pip install pandas scikit-learn matplotlib seaborn openpyxl pyarrow joblib xgboost -q

# ============================================================
# IMPORT LIBRARIES
# ============================================================

import pandas as pd
import numpy as np
import shap
import joblib

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans

# ============================================================
# UPLOAD DATASET
# ============================================================

print("\n====================================")
print("UPLOAD DATASET")
print("====================================")

from google.colab import files

uploaded = files.upload()

file_name = list(uploaded.keys())[0]

# ============================================================
# LOAD DATASET
# ============================================================

if file_name.endswith(".csv"):

    df = pd.read_csv(file_name)

elif file_name.endswith(".xlsx"):

    df = pd.read_excel(file_name)

else:

    raise ValueError("Unsupported File Format")

# ============================================================
# CLEAN COLUMN NAMES
# ============================================================

df.columns = df.columns.str.strip()

# ============================================================
# SHOW AVAILABLE COLUMNS
# ============================================================

print("\n====================================")
print("AVAILABLE COLUMNS")
print("====================================")

for col in df.columns:

    print(col)

# ============================================================
# TARGET COLUMN
# ============================================================

target_column = input(
    "\nEnter Target Column Name: "
).strip()

# ============================================================
# HANDLE MISSING VALUES
# ============================================================

df = df.fillna("Missing")

# ============================================================
# CONVERT EVERYTHING TO STRING
# ============================================================

for col in df.columns:

    df[col] = df[col].astype(str)

# ============================================================
# LABEL ENCODING
# ============================================================

label_encoders = {}

for col in df.columns:

    le = LabelEncoder()

    df[col] = le.fit_transform(df[col])

    label_encoders[col] = le

# ============================================================
# FEATURE TARGET SPLIT
# ============================================================

X = df.drop(columns=[target_column])

y = df[target_column]

# ============================================================
# DEFINE LEARNING KEYWORDS
# ============================================================

learning_keywords = [

    'study',
    'attendance',
    'assignment',
    'academic',
    'learning',
    'performance',
    'marks',
    'grade',
    'course',
    'exam'
]

# ============================================================
# DEFINE BEHAVIOR KEYWORDS
# ============================================================

behavior_keywords = [

    'stress',
    'sleep',
    'motivation',
    'social',
    'behavior',
    'emotion',
    'interaction',
    'anxiety',
    'participation',
    'mental'
]

# ============================================================
# AUTOMATIC FEATURE GROUPING
# ============================================================

learning_features = []
behavior_features = []

for col in X.columns:

    col_lower = col.lower()

    for keyword in learning_keywords:

        if keyword in col_lower:

            learning_features.append(col)

            break

    for keyword in behavior_keywords:

        if keyword in col_lower:

            behavior_features.append(col)

            break

# ============================================================
# FALLBACK FEATURES
# ============================================================

if len(learning_features) == 0:

    learning_features = list(X.columns[:5])

if len(behavior_features) == 0:

    behavior_features = list(X.columns[5:10])

# Remove duplicates
learning_features = list(set(learning_features))
behavior_features = list(set(behavior_features))

print("\n====================================")
print("LEARNING FEATURES")
print("====================================")

print(learning_features)

print("\n====================================")
print("BEHAVIOR FEATURES")
print("====================================")

print(behavior_features)

# ============================================================
# FEATURE SCALING
# ============================================================

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(
    X_scaled,
    columns=X.columns
)

# ============================================================
# CONTEXT FEATURE ENGINEERING
# ============================================================

def create_context_features(X):

    X_context = X.copy()

    X_context['engagement_index'] = X.mean(axis=1)

    X_context['motivation_score'] = X.std(axis=1)

    X_context['behavior_stability'] = X.var(axis=1)

    X_context['learning_consistency'] = (
        X.max(axis=1) - X.min(axis=1)
    )

    return X_context

X_context = create_context_features(
    X_scaled
)

# ============================================================
# SEQUENTIAL DEPENDENCY MODELING
# ============================================================

def create_sequential_features(X):

    X_seq = X.copy()

    for col in list(X.columns[:5]):

        X_seq[f"{col}_lag1"] = (
            X[col].shift(1)
        )

    X_seq = X_seq.fillna(0)

    return X_seq

X_seq = create_sequential_features(
    X_context
)

# ============================================================
# STUDENT-SPECIFIC REWARD ADAPTATION
# ============================================================

print("\n====================================")
print("CREATING PERSONALIZED STUDENT CLUSTERS")
print("====================================")

kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

student_clusters = kmeans.fit_predict(
    X_seq
)

X_seq['Student_Cluster'] = student_clusters

# ============================================================
# CLUSTER LABELS
# ============================================================

cluster_labels = {

    0: "Highly Motivated Learner",

    1: "Behaviorally Sensitive Learner",

    2: "Academically Stable Learner",

    3: "Low Engagement Learner"
}

# ============================================================
# TRAIN TEST SPLIT
# ============================================================

X_model = X_seq.drop(
    columns=['Student_Cluster']
)

X_train, X_test, y_train, y_test = train_test_split(

    X_model,
    y,

    test_size=0.2,

    random_state=42,

    stratify=y
)

# ============================================================
# MAIN REWARD MODEL
# ============================================================

reward_model = LogisticRegression(
    max_iter=1000
)

reward_model.fit(
    X_train,
    y_train
)

# ============================================================
# POLICY MODEL
# ============================================================

policy_model = RandomForestClassifier(

    n_estimators=300,

    random_state=42
)

policy_model.fit(
    X_train,
    y_train
)

# ============================================================
# PREDICTIONS
# ============================================================

y_pred = policy_model.predict(X_test)

# ============================================================
# PERFORMANCE
# ============================================================

accuracy = accuracy_score(
    y_test,
    y_pred
)

precision = precision_score(
    y_test,
    y_pred,
    average='weighted'
)

recall = recall_score(
    y_test,
    y_pred,
    average='weighted'
)

f1 = f1_score(
    y_test,
    y_pred,
    average='weighted'
)

print("\n====================================")
print("CASB-IRL PERFORMANCE")
print("====================================")

print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1:.4f}")

# ============================================================
# CLASSIFICATION REPORT
# ============================================================

print("\n====================================")
print("CLASSIFICATION REPORT")
print("====================================")

print(classification_report(
    y_test,
    y_pred
))

# ============================================================
# CONFUSION MATRIX
# ============================================================

cm = confusion_matrix(
    y_test,
    y_pred
)

plt.figure(figsize=(8,6))

sns.heatmap(

    cm,

    annot=True,

    fmt='d',

    cmap='Blues'
)

plt.title("CASB-IRL Confusion Matrix")

plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()

# ============================================================
# SHAP ANALYSIS
# ============================================================

print("\n====================================")
print("GENERATING SHAP ANALYSIS")
print("====================================")

explainer = shap.TreeExplainer(
    policy_model
)

shap_values = explainer.shap_values(
    X_test
)

shap.summary_plot(
    shap_values,
    X_test
)

# ============================================================
# DUAL REWARD MODELS
# ============================================================

learning_features_available = [

    col for col in learning_features

    if col in X_model.columns
]

behavior_features_available = [

    col for col in behavior_features

    if col in X_model.columns
]

# ============================================================
# LEARNING REWARD MODEL
# ============================================================

learning_reward_model = LogisticRegression(
    max_iter=1000
)

learning_reward_model.fit(

    X_train[learning_features_available],

    y_train
)

# ============================================================
# BEHAVIOR REWARD MODEL
# ============================================================

behavior_reward_model = LogisticRegression(
    max_iter=1000
)

behavior_reward_model.fit(

    X_train[behavior_features_available],

    y_train
)

# ============================================================
# PERSONALIZED REWARD MODELS
# ============================================================

print("\n====================================")
print("TRAINING PERSONALIZED REWARD MODELS")
print("====================================")

personalized_reward_models = {}

for cluster_id in np.unique(student_clusters):

    cluster_indices = np.where(
        student_clusters == cluster_id
    )[0]

    cluster_X = X_model.iloc[
        cluster_indices
    ]

    cluster_y = y.iloc[
        cluster_indices
    ]

    if len(cluster_X) < 5:

        continue

    model = LogisticRegression(
        max_iter=1000
    )

    model.fit(
        cluster_X,
        cluster_y
    )

    personalized_reward_models[
        cluster_id
    ] = model

    print(
        f"Cluster {cluster_id} trained"
    )

# ============================================================
# CAUSAL BEHAVIOR REASONING
# ============================================================

behavior_reasoning = {

    'stress': (
        "High stress reduces concentration and learning retention."
    ),

    'sleep': (
        "Poor sleep negatively affects memory consolidation."
    ),

    'motivation': (
        "Low motivation reduces engagement and learning persistence."
    ),

    'social': (
        "Excessive social activity diverts academic attention."
    ),

    'anxiety': (
        "Anxiety reduces confidence and classroom participation."
    ),

    'participation': (
        "Low participation weakens interactive learning."
    )
}

# ============================================================
# EXCEL REPORT STORAGE
# ============================================================

report_rows = []

# ============================================================
# LOCAL REWARD ATTRIBUTION
# ============================================================

def local_reward_attribution(student_index):

    student_features = X_seq.iloc[
        student_index
    ]

    # ========================================================
    # STUDENT PROFILE
    # ========================================================

    student_cluster = student_clusters[
        student_index
    ]

    cluster_name = cluster_labels.get(

        student_cluster,

        "General Learner"
    )

    # ========================================================
    # PERSONALIZED MODEL
    # ========================================================

    if student_cluster in personalized_reward_models:

        personalized_model = (

            personalized_reward_models[
                student_cluster
            ]
        )

        personalized_weights = (
            personalized_model.coef_[0]
        )

    else:

        personalized_weights = (
            reward_model.coef_[0]
        )

    # ========================================================
    # LEARNING REWARD
    # ========================================================

    learning_reward = np.dot(

        student_features[
            learning_features_available
        ],

        learning_reward_model.coef_[0]
    )

    # ========================================================
    # BEHAVIOR REWARD
    # ========================================================

    behavior_reward = np.dot(

        student_features[
            behavior_features_available
        ],

        behavior_reward_model.coef_[0]
    )

    # ========================================================
    # PERSONALIZED REWARD
    # ========================================================

    personalized_reward = np.dot(

        student_features.drop(
            labels=['Student_Cluster']
        ),

        personalized_weights
    )

    # ========================================================
    # COMPOSITE REWARD
    # ========================================================

    total_reward = (

        0.5 * learning_reward +

        0.3 * behavior_reward +

        0.2 * personalized_reward
    )

    # ========================================================
    # PREDICTION
    # ========================================================

    prediction = policy_model.predict(

        [student_features.drop(
            labels=['Student_Cluster']
        )]

    )[0]

    # ========================================================
    # LEARNING CONTRIBUTIONS
# ========================================================

    learning_contributions = pd.DataFrame({

        'Feature': learning_features_available,

        'Contribution':

        student_features[
            learning_features_available
        ].values *

        learning_reward_model.coef_[0]
    })

    # ========================================================
    # BEHAVIOR CONTRIBUTIONS
# ========================================================

    behavior_contributions = pd.DataFrame({

        'Feature': behavior_features_available,

        'Contribution':

        student_features[
            behavior_features_available
        ].values *

        behavior_reward_model.coef_[0]
    })

    # ========================================================
    # POSITIVE NEGATIVE FACTORS
# ========================================================

    positive_learning = learning_contributions[
        learning_contributions['Contribution'] > 0
    ]

    negative_learning = learning_contributions[
        learning_contributions['Contribution'] < 0
    ]

    positive_behavior = behavior_contributions[
        behavior_contributions['Contribution'] > 0
    ]

    negative_behavior = behavior_contributions[
        behavior_contributions['Contribution'] < 0
    ]

    # ========================================================
    # SORTING
# ========================================================

    positive_learning = positive_learning.sort_values(
        by='Contribution',
        ascending=False
    )

    negative_learning = negative_learning.sort_values(
        by='Contribution'
    )

    positive_behavior = positive_behavior.sort_values(
        by='Contribution',
        ascending=False
    )

    negative_behavior = negative_behavior.sort_values(
        by='Contribution'
    )

    # ========================================================
    # BEHAVIORAL EFFECT
# ========================================================

    if behavior_reward < 0:

        behavior_effect = (

            "Negative behavior reduces "
            "learning effectiveness"
        )

    else:

        behavior_effect = (

            "Positive behavior supports "
            "learning effectiveness"
        )

    # ========================================================
    # PERSONALIZED INTERPRETATION
# ========================================================

    if student_cluster == 1:

        personalization_text = (

            "Student is highly sensitive "
            "to behavioral fluctuations."
        )

    elif student_cluster == 3:

        personalization_text = (

            "Student shows low engagement "
            "patterns affecting learning."
        )

    elif student_cluster == 0:

        personalization_text = (

            "Student demonstrates strong "
            "motivational learning patterns."
        )

    else:

        personalization_text = (

            "Student exhibits stable "
            "academic behavior."
        )

    # ========================================================
    # CAUSAL EXPLANATIONS
# ========================================================

    causal_explanations = []

    for _, row in negative_behavior.head(5).iterrows():

        feature_name = row['Feature'].lower()

        for keyword, explanation in behavior_reasoning.items():

            if keyword in feature_name:

                causal_explanations.append(

                    f"{row['Feature']} : {explanation}"
                )

    # ========================================================
    # DISPLAY
# ========================================================

    print("\n====================================")
    print(f"STUDENT : {student_index}")
    print("====================================")

    print(f"\nPrediction : {prediction}")

    print(f"\nStudent Profile : {cluster_name}")

    print(f"\nLearning Reward : {learning_reward:.4f}")

    print(f"\nBehavior Reward : {behavior_reward:.4f}")

    print(f"\nPersonalized Reward : {personalized_reward:.4f}")

    print(f"\nComposite Reward : {total_reward:.4f}")

    print("\nBehavioral Effect:")

    print(behavior_effect)

    print("\nPersonalized Interpretation:")

    print(personalization_text)

    print("\n------------------------------------")
    print("WHY PERFORMANCE DEGRADATION OCCURRED")
    print("------------------------------------")

    for exp in causal_explanations:

        print(exp)

    # ========================================================
    # RETURN
# ========================================================

    return {

        'prediction': prediction,

        'cluster_name': cluster_name,

        'learning_reward': learning_reward,

        'behavior_reward': behavior_reward,

        'personalized_reward': personalized_reward,

        'total_reward': total_reward,

        'behavior_effect': behavior_effect,

        'personalization_text': personalization_text,

        'causal_explanations':

        ' | '.join(causal_explanations),

        'positive_learning':

        positive_learning.head(5),

        'negative_learning':

        negative_learning.head(5),

        'positive_behavior':

        positive_behavior.head(5),

        'negative_behavior':

        negative_behavior.head(5)
    }

# ============================================================
# GENERATE REPORT
# ============================================================

print("\n====================================")
print("GENERATING EXCEL REPORT")
print("====================================")

for student_index in range(len(X_seq)):

    result = local_reward_attribution(
        student_index
    )

    pos_learning = []
    neg_learning = []

    pos_behavior = []
    neg_behavior = []

    for _, row in result[
        'positive_learning'
    ].iterrows():

        pos_learning.append(

            f"{row['Feature']} "
            f"(+{row['Contribution']:.2f})"
        )

    for _, row in result[
        'negative_learning'
    ].iterrows():

        neg_learning.append(

            f"{row['Feature']} "
            f"({row['Contribution']:.2f})"
        )

    for _, row in result[
        'positive_behavior'
    ].iterrows():

        pos_behavior.append(

            f"{row['Feature']} "
            f"(+{row['Contribution']:.2f})"
        )

    for _, row in result[
        'negative_behavior'
    ].iterrows():

        neg_behavior.append(

            f"{row['Feature']} "
            f"({row['Contribution']:.2f})"
        )

    report_rows.append({

        'Student_ID': student_index,

        'Predicted_Learning_Type':

        result['prediction'],

        'Student_Profile':

        result['cluster_name'],

        'Learning_Reward':

        round(
            result['learning_reward'],
            4
        ),

        'Behavior_Reward':

        round(
            result['behavior_reward'],
            4
        ),

        'Personalized_Reward':

        round(
            result['personalized_reward'],
            4
        ),

        'Composite_Reward':

        round(
            result['total_reward'],
            4
        ),

        'Behavioral_Effect':

        result['behavior_effect'],

        'Personalized_Interpretation':

        result['personalization_text'],

        'Why_Performance_Degraded':

        result['causal_explanations'],

        'Positive_Learning_Factors':

        ', '.join(pos_learning),

        'Negative_Learning_Factors':

        ', '.join(neg_learning),

        'Positive_Behavior_Factors':

        ', '.join(pos_behavior),

        'Negative_Behavior_Factors':

        ', '.join(neg_behavior)
    })

# ============================================================
# REPORT DATAFRAME
# ============================================================

report_df = pd.DataFrame(
    report_rows
)

print("\n====================================")
print("REPORT PREVIEW")
print("====================================")

print(report_df.head())

# ============================================================
# SAVE EXCEL FILE
# ============================================================

excel_file = (
    "CASB_IRL_Personalized_Explainable_Report.xlsx"
)

report_df.to_excel(

    excel_file,

    index=False
)

# ============================================================
# DOWNLOAD REPORT
# ============================================================

from google.colab import files

files.download(excel_file)

print("\n====================================")
print("EXCEL REPORT DOWNLOADED")
print("====================================")

# ============================================================
# SAVE MODELS
# ============================================================

joblib.dump(

    policy_model,

    "casb_irl_policy_model.pkl"
)

joblib.dump(

    reward_model,

    "casb_irl_reward_model.pkl"
)

print("\n====================================")
print("MODELS SAVED")
print("====================================")
```
