```python
# ============================================================
# CASB-IRL COMPLETE FRAMEWORK
# WITH:
# ✔ Local Reward Attribution
# ✔ Dual Reward Context Modeling
# ✔ Learning vs Behavior Analysis
# ✔ Explainable Reward System
# ✔ Excel Report Generation
# ✔ Frontend Ready Output
# ✔ Google Colab Ready
# ============================================================

# ============================================================
# INSTALL REQUIRED LIBRARIES
# ============================================================

!pip uninstall -y shap -q

!pip install shap==0.46.0 -q
!pip install pandas scikit-learn matplotlib seaborn openpyxl pyarrow joblib -q

# ============================================================
# IMPORT LIBRARIES
# ============================================================

import pandas as pd
import numpy as np
import shap
import joblib

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# ============================================================
# UPLOAD DATASET
# ============================================================

print("\n====================================")
print("UPLOAD DATASET")
print("====================================")

from google.colab import files

uploaded = files.upload()

file_name = list(uploaded.keys())[0]

# ============================================================
# LOAD DATASET
# ============================================================

if file_name.endswith(".csv"):

    df = pd.read_csv(file_name)

elif file_name.endswith(".xlsx"):

    df = pd.read_excel(file_name)

else:

    raise ValueError("Unsupported File Format")

# ============================================================
# CLEAN COLUMN NAMES
# ============================================================

df.columns = df.columns.str.strip()

# ============================================================
# SHOW COLUMNS
# ============================================================

print("\n====================================")
print("AVAILABLE COLUMNS")
print("====================================")

for col in df.columns:

    print(col)

# ============================================================
# TARGET COLUMN
# ============================================================

target_column = input(
    "\nEnter Target Column Name: "
)

# ============================================================
# HANDLE MISSING VALUES
# ============================================================

df = df.fillna("Missing")

# ============================================================
# CONVERT ALL COLUMNS TO STRING
# ============================================================

for col in df.columns:

    df[col] = df[col].astype(str)

# ============================================================
# LABEL ENCODING
# ============================================================

label_encoders = {}

for col in df.columns:

    le = LabelEncoder()

    df[col] = le.fit_transform(df[col])

    label_encoders[col] = le

# ============================================================
# FEATURE TARGET SPLIT
# ============================================================

X = df.drop(columns=[target_column])

y = df[target_column]

# ============================================================
# DEFINE LEARNING FEATURES
# MODIFY ACCORDING TO YOUR DATASET
# ============================================================

learning_keywords = [

    'study',
    'attendance',
    'assignment',
    'academic',
    'learning',
    'performance',
    'marks',
    'grade',
    'course'
]

# ============================================================
# DEFINE BEHAVIOR FEATURES
# MODIFY ACCORDING TO YOUR DATASET
# ============================================================

behavior_keywords = [

    'stress',
    'sleep',
    'motivation',
    'social',
    'behavior',
    'emotion',
    'interaction',
    'anxiety',
    'participation'
]

# ============================================================
# AUTOMATIC FEATURE GROUPING
# ============================================================

learning_features = []

behavior_features = []

for col in X.columns:

    col_lower = col.lower()

    for keyword in learning_keywords:

        if keyword in col_lower:

            learning_features.append(col)

            break

    for keyword in behavior_keywords:

        if keyword in col_lower:

            behavior_features.append(col)

            break

# Remove duplicates
learning_features = list(set(learning_features))
behavior_features = list(set(behavior_features))

# ============================================================
# FALLBACK IF NO FEATURES FOUND
# ============================================================

if len(learning_features) == 0:

    learning_features = list(X.columns[:5])

if len(behavior_features) == 0:

    behavior_features = list(X.columns[5:10])

print("\n====================================")
print("LEARNING FEATURES")
print("====================================")

print(learning_features)

print("\n====================================")
print("BEHAVIOR FEATURES")
print("====================================")

print(behavior_features)

# ============================================================
# FEATURE SCALING
# ============================================================

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(

    X_scaled,

    columns=X.columns
)

# ============================================================
# CONTEXT FEATURE ENGINEERING
# ============================================================

def create_context_features(X):

    X_context = X.copy()

    X_context['engagement_index'] = X.mean(axis=1)

    X_context['motivation_score'] = X.std(axis=1)

    X_context['behavior_stability'] = X.var(axis=1)

    X_context['learning_consistency'] = (
        X.max(axis=1) - X.min(axis=1)
    )

    return X_context

X_context = create_context_features(
    X_scaled
)

# ============================================================
# SEQUENTIAL DEPENDENCY MODELING
# ============================================================

def create_sequential_features(X):

    X_seq = X.copy()

    for col in list(X.columns[:5]):

        X_seq[f"{col}_lag1"] = (
            X[col].shift(1)
        )

    X_seq = X_seq.fillna(0)

    return X_seq

X_seq = create_sequential_features(
    X_context
)

# ============================================================
# TRAIN TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(

    X_seq,
    y,

    test_size=0.2,

    random_state=42,

    stratify=y
)

# ============================================================
# MAIN REWARD MODEL
# ============================================================

reward_model = LogisticRegression(
    max_iter=1000
)

reward_model.fit(
    X_train,
    y_train
)

# ============================================================
# POLICY MODEL
# ============================================================

policy_model = RandomForestClassifier(

    n_estimators=300,

    random_state=42
)

policy_model.fit(
    X_train,
    y_train
)

# ============================================================
# PREDICTIONS
# ============================================================

y_pred = policy_model.predict(X_test)

# ============================================================
# PERFORMANCE
# ============================================================

accuracy = accuracy_score(
    y_test,
    y_pred
)

precision = precision_score(
    y_test,
    y_pred,
    average='weighted'
)

recall = recall_score(
    y_test,
    y_pred,
    average='weighted'
)

f1 = f1_score(
    y_test,
    y_pred,
    average='weighted'
)

print("\n====================================")
print("CASB-IRL PERFORMANCE")
print("====================================")

print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1:.4f}")

# ============================================================
# CONFUSION MATRIX
# ============================================================

cm = confusion_matrix(
    y_test,
    y_pred
)

plt.figure(figsize=(7,6))

sns.heatmap(

    cm,

    annot=True,

    fmt='d',

    cmap='Blues'
)

plt.title("CASB-IRL Confusion Matrix")

plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()

# ============================================================
# SHAP ANALYSIS
# ============================================================

print("\n====================================")
print("GENERATING SHAP ANALYSIS")
print("====================================")

explainer = shap.TreeExplainer(
    policy_model
)

shap_values = explainer.shap_values(
    X_test
)

shap.summary_plot(
    shap_values,
    X_test
)

# ============================================================
# DUAL REWARD MODELS
# ============================================================

learning_features_available = [

    col for col in learning_features

    if col in X_seq.columns
]

behavior_features_available = [

    col for col in behavior_features

    if col in X_seq.columns
]

# ============================================================
# LEARNING REWARD MODEL
# ============================================================

learning_reward_model = LogisticRegression(
    max_iter=1000
)

learning_reward_model.fit(

    X_train[learning_features_available],

    y_train
)

# ============================================================
# BEHAVIOR REWARD MODEL
# ============================================================

basis_reward_model = LogisticRegression(
    max_iter=1000
)

basis_reward_model.fit(

    X_train[behavior_features_available],

    y_train
)

# ============================================================
# EXCEL REPORT STORAGE
# ============================================================

report_rows = []

# ============================================================
# LOCAL REWARD ATTRIBUTION
# ============================================================

def local_reward_attribution(student_index):

    student_features = X_seq.iloc[student_index]

    # ========================================================
    # LEARNING REWARD
    # ========================================================

    learning_reward = np.dot(

        student_features[
            learning_features_available
        ],

        learning_reward_model.coef_[0]
    )

    # ========================================================
    # BEHAVIOR REWARD
    # ========================================================

    behavior_reward = np.dot(

        student_features[
            behavior_features_available
        ],

        basis_reward_model.coef_[0]
    )

    # ========================================================
    # COMPOSITE REWARD
    # ========================================================

    total_reward = (

        0.6 * learning_reward +

        0.4 * behavior_reward
    )

    # ========================================================
    # PREDICTION
    # ========================================================

    prediction = policy_model.predict(

        [student_features]

    )[0]

    # ========================================================
    # LEARNING CONTRIBUTIONS
    # ========================================================

    learning_contributions = pd.DataFrame({

        'Feature': learning_features_available,

        'Contribution':

        student_features[
            learning_features_available
        ].values *

        learning_reward_model.coef_[0]
    })

    # ========================================================
    # BEHAVIOR CONTRIBUTIONS
    # ========================================================

    behavior_contributions = pd.DataFrame({

        'Feature': behavior_features_available,

        'Contribution':

        student_features[
            behavior_features_available
        ].values *

        basis_reward_model.coef_[0]
    })

    # ========================================================
    # POSITIVE LEARNING FACTORS
    # ========================================================

    positive_learning = learning_contributions[
        learning_contributions['Contribution'] > 0
    ]

    # ========================================================
    # NEGATIVE LEARNING FACTORS
    # ========================================================

    negative_learning = learning_contributions[
        learning_contributions['Contribution'] < 0
    ]

    # ========================================================
    # POSITIVE BEHAVIOR FACTORS
    # ========================================================

    positive_behavior = behavior_contributions[
        behavior_contributions['Contribution'] > 0
    ]

    # ========================================================
    # NEGATIVE BEHAVIOR FACTORS
    # ========================================================

    negative_behavior = behavior_contributions[
        behavior_contributions['Contribution'] < 0
    ]

    # ========================================================
    # SORT
    # ========================================================

    positive_learning = positive_learning.sort_values(

        by='Contribution',

        ascending=False
    )

    negative_learning = negative_learning.sort_values(

        by='Contribution'
    )

    positive_behavior = positive_behavior.sort_values(

        by='Contribution',

        ascending=False
    )

    negative_behavior = negative_behavior.sort_values(

        by='Contribution'
    )

    # ========================================================
    # INTERPRETATION
    # ========================================================

    if behavior_reward < 0:

        behavior_effect = (

            "Negative behavior is reducing "
            "learning effectiveness"
        )

    else:

        behavior_effect = (

            "Positive behavior supports "
            "learning effectiveness"
        )

    # ========================================================
    # DISPLAY
    # ========================================================

    print("\n====================================")
    print(f"STUDENT : {student_index}")
    print("====================================")

    print(f"\nPrediction : {prediction}")

    print(f"\nLearning Reward : {learning_reward:.4f}")

    print(f"\nBehavior Reward : {behavior_reward:.4f}")

    print(f"\nComposite Reward : {total_reward:.4f}")

    print("\nBehavioral Effect:")

    print(behavior_effect)

    print("\n------------------------------------")
    print("TOP NEGATIVE BEHAVIOR FACTORS")
    print("------------------------------------")

    print(negative_behavior.head())

    # ========================================================
    # RETURN DATA
    # ========================================================

    return {

        'prediction': prediction,

        'learning_reward': learning_reward,

        'behavior_reward': behavior_reward,

        'total_reward': total_reward,

        'behavior_effect': behavior_effect,

        'positive_learning':

        positive_learning.head(5),

        'negative_learning':

        negative_learning.head(5),

        'positive_behavior':

        positive_behavior.head(5),

        'negative_behavior':

        negative_behavior.head(5)
    }

# ============================================================
# GENERATE REPORT FOR ALL STUDENTS
# ============================================================

print("\n====================================")
print("GENERATING EXCEL REPORT")
print("====================================")

for student_index in range(len(X_seq)):

    result = local_reward_attribution(
        student_index
    )

    # ========================================================
    # POSITIVE LEARNING
    # ========================================================

    pos_learning = []

    for _, row in result[
        'positive_learning'
    ].iterrows():

        pos_learning.append(

            f"{row['Feature']} "

            f"(+{row['Contribution']:.2f})"
        )

    # ========================================================
    # NEGATIVE LEARNING
    # ========================================================

    neg_learning = []

    for _, row in result[
        'negative_learning'
    ].iterrows():

        neg_learning.append(

            f"{row['Feature']} "

            f"({row['Contribution']:.2f})"
        )

    # ========================================================
    # POSITIVE BEHAVIOR
    # ========================================================

    pos_behavior = []

    for _, row in result[
        'positive_behavior'
    ].iterrows():

        pos_behavior.append(

            f"{row['Feature']} "

            f"(+{row['Contribution']:.2f})"
        )

    # ========================================================
    # NEGATIVE BEHAVIOR
    # ========================================================

    neg_behavior = []

    for _, row in result[
        'negative_behavior'
    ].iterrows():

        neg_behavior.append(

            f"{row['Feature']} "

            f"({row['Contribution']:.2f})"
        )

    # ========================================================
    # ADD ROW
    # ========================================================

    report_rows.append({

        'Student_ID': student_index,

        'Predicted_Learning_Type':

        result['prediction'],

        'Learning_Reward':

        round(
            result['learning_reward'],
            4
        ),

        'Behavior_Reward':

        round(
            result['behavior_reward'],
            4
        ),

        'Composite_Reward':

        round(
            result['total_reward'],
            4
        ),

        'Behavioral_Effect_On_Learning':

        result['behavior_effect'],

        'Positive_Learning_Factors':

        ', '.join(pos_learning),

        'Negative_Learning_Factors':

        ', '.join(neg_learning),

        'Positive_Behavior_Factors':

        ', '.join(pos_behavior),

        'Negative_Behavior_Factors':

        ', '.join(neg_behavior)
    })

# ============================================================
# CREATE REPORT DATAFRAME
# ============================================================

report_df = pd.DataFrame(report_rows)

print("\n====================================")
print("REPORT PREVIEW")
print("====================================")

print(report_df.head())

# ============================================================
# SAVE EXCEL FILE
# ============================================================

excel_file = (
    "CASB_IRL_Explainable_Reward_Report.xlsx"
)

report_df.to_excel(

    excel_file,

    index=False
)

# ============================================================
# DOWNLOAD EXCEL FILE
# ============================================================

from google.colab import files

files.download(excel_file)

print("\n====================================")
print("EXCEL FILE DOWNLOADED")
print("====================================")

# ============================================================
# SAVE MODEL
# ============================================================

joblib.dump(

    policy_model,

    "casb_irl_model.pkl"
)

print("\n====================================")
print("MODEL SAVED")
print("====================================")
```
